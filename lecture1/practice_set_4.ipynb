### Question 1

Given: s = "python programming"

### Question 2

Given: s = " Hello World "

Task: Remove spaces on both sides and replace the space between words with - → "Hello-World"

```python
s = " Hello World "
ns = s.strip()
r = ns.replace(" ", "-")
print(ns)
print(r)
```

### Question 3

Given: s = "banana"

Task: Replace only the last occurrence of "a" with "A" without using loops.

```python
s = "banana"

index = s.rfind("a")
result = s[:index] + "A" + s[index+1:]

print(result)
```

### Question 6

Given: s = "I love python programming"

Task: Using slicing, reverse only the substring "python" while keeping the rest same.

Expected Output → "I love nohtyp programming"

```python
s = "I love python programming"
s1 = s.index("python")
s2 = s1 + len("python")
str = s[:s1] + s[s1 : s2][::-1] + s[s2 :]
print(str)
```
